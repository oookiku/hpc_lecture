```python
import numpy as np
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata
%matplotlib inline

nx = 41
ny = 41

x_tmp = np.empty(nx*ny)
y_tmp = np.empty(nx*ny)
u_tmp = np.empty(nx*ny)
v_tmp = np.empty(nx*ny)
p_tmp = np.empty(nx*ny)
```

## When nt = 100

```python
f = open('result1.dat', 'r')
line = f.readline()

i = 0
while line:
    data = line.split()
    x_tmp[i] = data[0]
    y_tmp[i] = data[1]
    u_tmp[i] = data[2]
    v_tmp[i] = data[3]
    p_tmp[i] = data[4]
    line = f.readline()
    i += 1
f.close()

x = x_tmp.reshape(ny,nx)
y = y_tmp.reshape(ny,nx)
u = u_tmp.reshape(ny,nx)
v = v_tmp.reshape(ny,nx)
p = p_tmp.reshape(ny,nx)

#
# the following codes are same as the distributed sample
#
fig = pyplot.figure(figsize=(11,7), dpi=100)
# plotting the pressure field as a contour
pyplot.contourf(x, y, p, alpha=0.5, cmap=cm.viridis)  
pyplot.colorbar()
# plotting the pressure field outlines
pyplot.contour(x, y, p, cmap=cm.viridis)  
# plotting velocity field
pyplot.quiver(x[::2, ::2], y[::2, ::2], u[::2, ::2], v[::2, ::2]) 
pyplot.xlabel('X')
pyplot.ylabel('Y');
```
